```python
fake = Faker()
Faker.seed(42)
random.seed(42)

# Data volume — enterprise scale
NUM_DENTISTS   = 50
NUM_PATIENTS   = 5000
NUM_TREATMENTS = 15000

print("Libraries loaded")
print(f"Generating: {NUM_DENTISTS} dentists, {NUM_PATIENTS} patients, {NUM_TREATMENTS} treatments")
```

## 1. Generate Dentists

```python
SPECIALTIES = [
    'General Dentistry', 'Endodontics', 'Periodontics',
    'Oral Surgery', 'Orthodontics', 'Pediatric Dentistry',
    'Prosthodontics', 'Cosmetic Dentistry'
]

dentists = []
for i in range(NUM_DENTISTS):
    dentists.append({
        'DENTIST_ID':     str(uuid.uuid4()),
        'FIRST_NAME':     fake.first_name(),
        'LAST_NAME':      fake.last_name(),
        'LICENSE_NUMBER': f'CA-{random.randint(10000,99999)}-{i}',
        'SPECIALTY':      random.choice(SPECIALTIES),
        'PHONE':          fake.numerify('###-###-####'),
        'EMAIL':          fake.email()
    })

df_dentists = pd.DataFrame(dentists)
print(f"Dentists generated: {len(df_dentists)}")
df_dentists.head()
```

## 2. Generate Insurance Companies

```python
INSURANCE_NAMES = [
    'Delta Dental', 'Cigna Dental', 'Aetna Dental',
    'MetLife Dental', 'Guardian Dental', 'United Concordia',
    'Humana Dental', 'Ameritas', 'Sun Life Dental', 'Principal Dental'
]

insurers = []
for name in INSURANCE_NAMES:
    insurers.append({
        'INSURANCE_COMPANY_ID':  str(uuid.uuid4()),
        'COMPANY_NAME':          name,
        'CONTACT_PHONE':         fake.numerify('###-###-####'),
        'CONTACT_EMAIL':         f'claims@{name.lower().replace(" ","")}.com',
        'SUBMISSION_PORTAL_URL': f'https://claims.{name.lower().replace(" ","")}.com',
        'AVG_PROCESSING_DAYS':   random.randint(7, 21),
        'REQUIRED_DOCUMENTS':    json.dumps(['xray', 'doctor_note']),
        'COVERAGE_RULES':        json.dumps({'max_annual': 2000, 'deductible': 50})
    })

df_insurers = pd.DataFrame(insurers)
print(f"Insurance companies generated: {len(df_insurers)}")
df_insurers.head()
```

## 3. Generate Patients

```python
insurance_ids = df_insurers['INSURANCE_COMPANY_ID'].tolist()

patients = []
for _ in range(NUM_PATIENTS):
    age = int(random.gauss(45, 15))
    age = max(18, min(80, age))
    dob = date.today() - timedelta(days=age*365)
    patients.append({
        'PATIENT_ID':             str(uuid.uuid4()),
        'FIRST_NAME':             fake.first_name(),
        'LAST_NAME':              fake.last_name(),
        'DATE_OF_BIRTH':          dob.isoformat(),
        'PHONE':                  fake.numerify('###-###-####'),
        'EMAIL':                  fake.email(),
        'ADDRESS':                fake.address().replace('\n', ', '),
        'INSURANCE_COMPANY_ID':   random.choice(insurance_ids),
        'INSURANCE_MEMBER_ID':    f'MBR{random.randint(100000,999999)}',
        'INSURANCE_GROUP_NUMBER': f'GRP{random.randint(1000,9999)}',
        'COVERAGE_DETAILS':       json.dumps({'coverage_pct': random.choice([70, 80, 90])})
    })

df_patients = pd.DataFrame(patients)
print(f"Patients generated: {len(df_patients)}")
df_patients.head()
```

## 4. Generate Treatments

```python
CDT_CODES = [
    {'code': 'D0120', 'type': 'Periodic Oral Examination',    'fee': 65},
    {'code': 'D0210', 'type': 'Full Mouth X-Ray',             'fee': 150},
    {'code': 'D1110', 'type': 'Adult Teeth Cleaning',         'fee': 120},
    {'code': 'D2140', 'type': 'Amalgam Filling - 1 Surface',  'fee': 180},
    {'code': 'D2750', 'type': 'Porcelain Crown',              'fee': 1400},
    {'code': 'D3310', 'type': 'Root Canal - Anterior',        'fee': 900},
    {'code': 'D3330', 'type': 'Root Canal - Molar',           'fee': 1200},
    {'code': 'D4341', 'type': 'Periodontal Scaling',          'fee': 280},
    {'code': 'D7140', 'type': 'Simple Tooth Extraction',      'fee': 200},
    {'code': 'D7240', 'type': 'Surgical Extraction - Wisdom', 'fee': 450},
]

patient_ids = df_patients['PATIENT_ID'].tolist()
dentist_ids = df_dentists['DENTIST_ID'].tolist()

treatments = []
for _ in range(NUM_TREATMENTS):
    cdt = random.choice(CDT_CODES)
    fee_variation = random.uniform(0.9, 1.1)
    treatments.append({
        'TREATMENT_ID':          str(uuid.uuid4()),
        'PATIENT_ID':            random.choice(patient_ids),
        'DENTIST_ID':            random.choice(dentist_ids),
        'TREATMENT_DATE':        fake.date_between('-3y', 'today').isoformat(),
        'TOOTH_NUMBER':          str(random.randint(1, 32)),
        'TREATMENT_TYPE':        cdt['type'],
        'TREATMENT_DESCRIPTION': f"{cdt['type']} performed on tooth {random.randint(1,32)}",
        'CDT_CODE':              cdt['code'],
        'FEE_CHARGED':           round(cdt['fee'] * fee_variation, 2),
        'DOCTOR_NOTES':          fake.sentence(nb_words=10)
    })

df_treatments = pd.DataFrame(treatments)
print(f"Treatments generated: {len(df_treatments)}")
df_treatments.head()
```

## 5. Generate Documents

```python
DOCUMENT_TYPES_BY_CDT = {
    'D2750': ['xray', 'photo', 'doctor_note'],
    'D3310': ['xray', 'doctor_note'],
    'D3330': ['xray', 'doctor_note'],
    'D7240': ['xray', 'photo'],
    'D4341': ['xray', 'doctor_note'],
    'D0120': ['doctor_note'],
    'D0210': ['xray'],
    'D1110': ['doctor_note'],
    'D2140': ['doctor_note'],
    'D7140': ['xray'],
}

documents = []
for _, t in df_treatments.iterrows():
    doc_types = DOCUMENT_TYPES_BY_CDT.get(t['CDT_CODE'], ['doctor_note'])
    for doc_type in doc_types:
        documents.append({
            'DOCUMENT_ID':   str(uuid.uuid4()),
            'TREATMENT_ID':  t['TREATMENT_ID'],
            'PATIENT_ID':    t['PATIENT_ID'],
            'DOCUMENT_TYPE': doc_type,
            'FILE_URL':      f'https://storage.dental-ai.com/{doc_type}/{uuid.uuid4()}.pdf'
        })

df_documents = pd.DataFrame(documents)
print(f"Documents generated: {len(df_documents)}")
df_documents.head()
```

## 7. Save to CSV

```python
os.makedirs('../data/synthetic', exist_ok=True)

df_dentists.to_csv('../data/synthetic/dentists.csv',     index=False)
df_insurers.to_csv('../data/synthetic/insurers.csv',     index=False)
df_patients.to_csv('../data/synthetic/patients.csv',     index=False)
df_treatments.to_csv('../data/synthetic/treatments.csv', index=False)
df_documents.to_csv('../data/synthetic/documents.csv',   index=False)

print("All CSV files saved:")
print(f"  Dentists:    {len(df_dentists)}")
print(f"  Insurers:    {len(df_insurers)}")
print(f"  Patients:    {len(df_patients)}")
print(f"  Treatments:  {len(df_treatments)}")
print(f"  Documents:   {len(df_documents)}")
```
